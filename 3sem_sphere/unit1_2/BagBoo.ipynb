```python
import matplotlib.pyplot as plt
import numpy as np
from utils import DesTreeReg, MyGradientBoostingRegressor, BagBoo, parse_sparce

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
X_train, y_train = parse_sparce('reg.train.txt')
X_test, y_test  = parse_sparce('reg.test.txt')
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
myBagBoo = BagBoo(base_estimator='MyGB', subsampleGB=0.3, subsampleBB = 0.3)
myBagBoo.fit(X_train, y_train, X_test, y_test)
```

```python
sk_BagBoo = BagBoo(base_estimator='GB', subsampleGB=0.3, subsampleBB = 0.3)
sk_BagBoo.fit(X_train, y_train, X_test, y_test)
```

```python
plt.plot(myBagBoo.mse_train, label = 'myBB mse train')
plt.plot(myBagBoo.mse_test, label = 'myBB mse test')
plt.plot(sk_BagBoo.mse_train, label = 'skBB mse train')
plt.plot(sk_BagBoo.mse_train * 0.97, linestyle='--')
plt.plot(sk_BagBoo.mse_train * 1.03, linestyle='--')
plt.plot(sk_BagBoo.mse_test, label = 'skBB mse test')
plt.plot(sk_BagBoo.mse_test * 0.97, linestyle='--')
plt.plot(sk_BagBoo.mse_test * 1.03, linestyle='--')
plt.legend()
```
