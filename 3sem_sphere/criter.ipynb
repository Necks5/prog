```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
split1 = []
was_clicked1 = []
with open('c1.txt') as f:
    for line in f:
        temp = list(map(int, line.split()))
        split1.append(temp)
        was_clicked1.append(max(temp))
```

```python
split2 = []
was_clicked2 = []
with open('c2.txt') as f:
    for line in f:
        temp = list(map(int, line.split()))
        split2.append(temp)
        was_clicked2.append(max(temp))
```

```python
split1 = np.array(split1)
split2 = np.array(split2)
was_clicked1 = np.array(was_clicked1)
was_clicked2 = np.array(was_clicked2)
```

```python
ctr1 = np.sum(was_clicked1.reshape(1000, 1000), axis = 1) / 1000
ctr2 = np.sum(was_clicked2.reshape(1000, 1000), axis = 1) / 1000
```

```python
plt.plot(ctr1)
plt.plot(ctr2)
```

```python
N = 100
count = 0
for i in range(N):
    np.random.shuffle(was_clicked1)
    np.random.shuffle(was_clicked2)
    wc1 = np.sum(was_clicked1.reshape(1000, 1000), axis = 1) / 1000
    wc2 = np.sum(was_clicked2.reshape(1000, 1000), axis = 1) / 1000
    t = np.abs(np.mean(wc1) - np.mean(wc2)) / np.sqrt((np.std(wc1)**2 + np.std(wc2)**2) / 1000)
    if t > 1.96:
        count += 1
```

```python
count
```

Вывод: нулевая гипотеза отвергается (в принципе отличия видны на графике)
